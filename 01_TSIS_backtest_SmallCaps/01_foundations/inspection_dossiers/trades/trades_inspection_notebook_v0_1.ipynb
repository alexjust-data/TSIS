## Como debe leerse este notebook

En `trades` hay un error de lectura muy facil de cometer: ver muchos conflictos contra referencias y concluir demasiado rapido que casi todo el tape esta roto.

Este notebook existe para evitar precisamente ese error.

La regla correcta es separar siempre:

1. poblacion agregada;
2. muestra metodologica de `380` casos;
3. buckets o familias semanticas;
4. y estados finales de certificacion.

Si mezclas esos cuatro niveles, acabas leyendo como `bad_data` lo que en realidad puede ser:

- conflicto de escala contra arbitros;
- conflicto de microestructura;
- falta de arbitro `1m`;
- o residuo `review` todavia pendiente de rehabilitacion.

## Que significa cada familia principal

### `reference_scale_mismatch`

No significa "file raro" ni "tape roto" por si solo.

Significa que el conflicto dominante es de **comparabilidad por escala** frente a arbitros como `daily` o `1m`. La pregunta correcta no es si el caso es feo, sino si vive en una escala distinta de la referencia.

### `review_microstructure`

No significa solo "hay review".

Significa que el dano vive en la **textura fina del tape**:

- odd-lots,
- duplicacion,
- bursts por timestamp,
- o comparabilidad fina del flujo.

### `review_no_1m_reference`

No significa que el caso este limpio.

Significa que hay conflicto, pero falta el arbitro `1m` para cerrarlo con precision. Es un estado de incertidumbre disciplinada, no una absolucion.

### `review_1m_reference_alignment`

No significa simple residuo menor.

Significa que al abrir el arbitro `1m` cambia la verdad del caso. Casos que parecian razonables en una vista mas gruesa dejan de serlo.

### `review`

No es un cajon vacio. Es el residuo que todavia no cae limpiamente en una familia mas especifica y por eso no debe promocionarse ni a `good` ni a `bad` sin una regla de rehabilitacion.

### `bad_data`

No significa solo "caso extremo visualmente feo".

Significa frontera semantica donde el flujo deja de ser economicamente defendible como tape de ejecucion o como comparacion con arbitros.

### `good`

No mide toda la masa util del bloque.

Mide la cola **pristine**, donde `trades`, `daily` y `1m` alinean casi impecablemente. Es muy pequeno a proposito, porque el umbral para entrar ahi es muy estricto.

## Como usar el selector

El selector tiene tres modos de lectura complementarios:

- `population`
  - para ver las imagenes agregadas y entender cuanta masa hay y de que tipo;
- `case_pack`
  - para ver casos historicos curados y altamente explicativos;
- `muestra_380`
  - para recorrer la muestra metodologica del notebook historico, mucho mas amplia que los ejemplos curados y ya con panel grafico del caso.

La regla practica es:

- usa `population` para entender el tamano del problema;
- usa `case_pack` para fijar la firma visual de cada familia;
- usa `muestra_380` para comprobar que la familia no depende de dos ejemplos bonitos, sino que se sostiene en una muestra mucho mas amplia con panel comparable.
